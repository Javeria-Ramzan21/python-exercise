```python
# Q1. Write a function that takes a string and returns:


def string_manipulator(s):
    reversed_str = s[::-1]
    every_second = s[::2]
    return reversed_str, every_second


rev, every_sec = string_manipulator("Hello World")
print("Q1 - Original:", "Hello World")
print("Q1 - Reversed:", rev)
print("Q1 - Every 2nd char:", every_sec)
print("-" * 40)
```

```python
# Q2. Given a list of numbers, return a new list containing


def squares_of_evens(numbers):
    return [num**2 for num in numbers if num % 2 == 0]


numbers = [1, 2, 3, 4, 5, 6, 7, 8]
print("Q2 - Original list:", numbers)
print("Q2 - Squares of evens:", squares_of_evens(numbers))
print("-" * 40)
```

```python
# Q3. Write a function that counts frequency of each character


def char_frequency(s):
    freq = {}
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    return freq


test_string = "hello"
print("Q3 - String:", test_string)
print("Q3 - Character frequency:", char_frequency(test_string))
print("-" * 40)
```

```python
# Q4. Write a function that takes a list and returns indices


def odd_indices(nums):
    return [idx for idx, num in enumerate(nums) if num % 2 != 0]


nums = [10, 15, 20, 25, 30, 35, 40]
print("Q4 - List:", nums)
print("Q4 - Indices of odd numbers:", odd_indices(nums))
print("-" * 40)
```

```python
# Q5. Write a function to rotate a list right by k positions using slicing

def rotate_list(lst, k):
    if not lst:
        return lst
    k = k % len(lst)
    return lst[-k:] + lst[:-k]


my_list = [1, 2, 3, 4, 5]
k = 2
print("Q5 - Original list:", my_list)
print("Q5 - Rotate right by", k, "positions:", rotate_list(my_list, k))
print("-" * 40)
```

```python
# Q6. Write a function that takes two lists and returns a dictionary using zip()

def lists_to_dict(keys, values):
    return dict(zip(keys, values))


keys = ['name', 'age', 'city']
values = ['Alice', 25, 'New York']
print("Q6 - Keys:", keys)
print("Q6 - Values:", values)
print("Q6 - Dictionary:", lists_to_dict(keys, values))
print("-" * 40)
```

```python
# Q7. Using iter() and next(), manually iterate through a list


def manual_iteration(lst):
    my_iter = iter(lst)
    while True:
        try:
            element = next(my_iter)
            print(element)
        except StopIteration:
            print("Reached the end of the list!")
            break


print("Q7 - Manual iteration of [10, 20, 30, 40]:")
manual_iteration([10, 20, 30, 40])
print("-" * 40)
```

```python
# Q8. Write a function that flattens a nested list using list comprehension

def flatten_list(nested_list):
    return [item for sublist in nested_list for item in sublist]


nested = [[1, 2], [3, 4], [5, 6, 7]]
print("Q8 - Nested list:", nested)
print("Q8 - Flattened list:", flatten_list(nested))
print("-" * 40)
```

```python
# Q9. Write a function that checks whether a string is a palindrome using slicing

def is_palindrome(s):
    s = s.replace(" ", "").lower()
    return s == s[::-1]


words = ["racecar", "hello", "A man a plan a canal panama", "python"]
print("Q9 - Palindrome checks:")
for word in words:
    print(f"  '{word}' -> {is_palindrome(word)}")
print("-" * 40)
```

```python
# Q10. Write a function that returns the second largest number in a list


def second_largest(nums):
    if len(nums) < 2:
        return None
    largest = second = float('-inf')
    for num in nums:
        if num > largest:
            second = largest
            largest = num
        elif num > second and num != largest:
            second = num
    return second if second != float('-inf') else None


numbers_list = [5, 2, 8, 1, 9, 3]
print("Q10 - List:", numbers_list)
print("Q10 - Second largest:", second_largest(numbers_list))
print("-" * 40)
```
